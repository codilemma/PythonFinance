```python
# The DataFrame Class
import pandas as pd

df = pd.DataFrame(
        [10,20,30,40],
        columns=['numbers'],
        index=['a','b','c','d'])
df
```

```python
# The 'index' attribute and 'Index' object
df.index
```

```python
# The 'columns' attribute and 'Index' object.
df.columns
```

```python
# Selects the value corresponding to index 'c'
df.loc['c']
```

```python
#'Selects the two values corresponding to indices 'a' and 'd'
df.loc[['a','d']]
```

```python
# Selects the second and third rows via the index positions.
df.iloc[1:3]
```

```python
# Calculates the sum of the single column
df.sum()
```

```python
# Use the 'apply()' method to calculate squares in vectorized fashion
df.apply(lambda x: x **2)
```

```python
# Applies vectorization directly as with 'ndarray' objects
df ** 2
```

```python
# Unlike 'NumPy ndarray' ojbects, we have both dimensions.abs
# Create a new column based on a Data Frame object.abs
df['floats'] = (1.5, 2.5, 3.5, 4.5)
df
```

```python
# Use whole DataFrame object to define a new column
df['names'] = pd.DataFrame(['Yves','Sandra','Lilli','Henry'],
                            index=['d','a','b','c'])

df
```

```python
# 1 - Appends a new row via a 'dict' object;
#     Temporary operation during which index information gets lost.
df.append({'numbers': 100, 'floats': 5.75, 'names': 'Jil'},
            ignore_index=True)
```

```python
# 2 - Appends row based on 'DataFrame' object with index info.
#     Original index is preserved
df = df.append(pd.DataFrame({'numbers': 100, 
                             'floats': 5.75,
                             'names':'Jil'},
                             index=['y']))

df
```

```python
# 3- Appends incomplete data row to 'DataFrame' object.  Fills with NaN 
#    vals
df = df.append(pd.DataFrame({'names': 'Liz'}, index=['z']), sort=False)
df
```

```python
# 4 - Returns different 'dtypes' of the single columns.
df.dtypes
```

```python
# 1- Calculates mean over the two specified columns
df[['numbers', 'floats']].mean()
```

```python
# 2 - Calculates the standard deviation over specified columns
df[['numbers','floats']].std()
```
